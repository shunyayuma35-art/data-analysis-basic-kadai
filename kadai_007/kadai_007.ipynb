```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
# 注文データ
df = pd.DataFrame({
    '商品番号': [1, 1, 2, 2, 3, 3, 3],
    '注文数': [5, 3, 2, 4, 1, 2, 3]
})

# 商品マスタ（カテゴリー）
category_df = pd.DataFrame({
    '商品番号': [1, 2, 3],
    'カテゴリー': ['弁当', 'デザート', '飲み物']
})
```

```python
df = pd.merge(
    df,
    category_df[['商品番号', 'カテゴリー']],
    how='inner',
    on='商品番号'
)
```

```python
df['カテゴリー'].value_counts().plot(kind='bar')
plt.show()
```

```python
df.groupby('商品番号')['注文数'].describe()
```
